### **File Context Managers:**

````python
#Using context managers (with statement) is a safer way to handle file I/O,
#as it automatically closes the file when you're done:
with open("filename.txt", "r") as file:
    content = file.read()

with open("filename.txt", "w") as file:
    file.write("New content using context manager.")

````

````python
#Reading and Writing Binary Files:

#For binary files (e.g., images, videos), use the "rb" mode for reading and "wb" mode for writing:

# Reading binary
with open("image.jpg", "rb") as file:
    image_data = file.read()

# Writing binary
with open("new_image.jpg", "wb") as file:
    file.write(image_data)
````

````python
with open("image.jpg", "rb") as file:
    image_data = file.read()
print(image_data)
````

## **Modules and Packages:**

**Module:**
A module is a single Python file that contains variables, functions, and classes that are related to a specific purpose. It's a way to organize code and make it reusable. You can use modules to encapsulate functionality, and then import and use them in other parts of your code. To create a module, you simply write your code in a .py file.

**Package:**
A package is a collection of related Python modules grouped together in a directory hierarchy. It allows you to organize your code in a more structured manner, particularly when your project becomes more complex. A package contains a special file named __init__.py, which is executed when the package is imported. Packages can contain sub-packages and modules.

### Creating your own module
**Create a Python File:**
Start by creating a new .py file with a name that you want to use as the module name. For example, let's create a module named mymodule.py.

**Write Code:**
In the mymodule.py file, write the code that you want to include in your module. This can be functions, variables, or classes.

````python
# mymodule.py
def greet(name):
    return f"Hello, {name}!"

PI = 3.14159
````

**Use the Module:**
Now you can use the mymodule module in other Python scripts. Place the mymodule.py file in the same directory as your script or in a directory that's in your Python path.

````python
import mymodule

print(mymodule.greet("Alice"))
print(mymodule.PI)
````

## **Creating a package**

Creating a package involves organizing multiple related modules into a directory hierarchy with an __init__.py file. This hierarchy allows you to group and manage your code more effectively.

Create a Directory:
Start by creating a new directory with the desired package name. Inside this directory, you can include multiple .py files, sub-packages, and an __init__.py file.

Let's create a package named my_package.

**Create Module Files:**

Inside the my_package directory, create the module files you want to include in the package. These module files will contain the code you want to include.

````python
my_package/
├── __init__.py
├── module1.py
└── module2.py
````

````python
#Write Code:
#Write your code in the module files.
#These can include functions, classes, and variables that you want to make available in your package.

# module1.py
def function1():
    print("Function 1")

# module2.py
def function2():
    print("Function 2")
````

**Create the __init__.py File:**
The __init__.py file is executed when the package is imported. You can use it to specify which modules and sub-packages to include in the package's namespace.

````python
# __init__.py
from . import module1
from . import module2
````

You can now use your package in other scripts by importing the modules and sub-packages.

````python
import my_package.module1
import my_package.module2

my_package.module1.function1()
my_package.module2.function2()

#Alternatively, you can use relative imports within the package:
from . import module1
from . import module2
````

## **Map() Function:**

The map() function in Python is used to apply a given function to all the elements of an iterable (such as a list, tuple, or string) and returns an iterator containing the results. It essentially applies the provided function to each element of the iterable and collects the results in a new iterable.

````python
#syntax
map(function, iterable)
````

````python
import math
m=map(math.sqrt,(9,16,25))
print(m)
print(list(m))
````

````python
# Define a function to find square root of  a number
def sqrt(x):
    return x ** 0.5

# List of numbers
numbers = [9, 16, 25]

# Apply the square root function to each element using map
squared_numbers = map(sqrt, numbers)

# Convert the map object to a list
squared_numbers_list = list(squared_numbers)

print(squared_numbers_list)
````

## **Zip() Function:**

The zip() function in Python is used to combine multiple iterables (e.g., lists, tuples) element-wise. It takes iterables as arguments and returns an iterator that generates tuples containing elements from the input iterables at the same index.

````python
#syntax
zip(iterable1, iterable2, ...)
````

````python
# Lists of names and ages
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 22]

# Zip the names and ages together
zipped_data = zip(names, ages)

# Convert the zip object to a list of tuples
zipped_data_list = list(zipped_data)
print(zipped_data)
print(zipped_data_list)
````
